```python
import pandas as pd
import numpy as np
```

```python
movies = pd.read_csv('movies.csv',sep=',')
rating = pd.read_csv('ratings.csv',sep=',')
```

```python
df = pd.merge(rating,movies,on='movieId')
```

```python
tabela_filmes = pd.pivot_table(df, index='title', columns='userId',values='rating').fillna(0)
```

```python
import sklearn.metrics.pairwise as pw
```

```python
rec = pw.cosine_similarity(tabela_filmes)
rec
```

```python
rec_df = pd.DataFrame(rec, columns=tabela_filmes.index, index=tabela_filmes.index)
rec_df.head(3)
```

```python
cossine_df = pd.DataFrame(rec_df['X-Men: First Class (2011)'].sort_values(ascending=False))
cossine_df.columns = ['Recomendações']
cossine_df.head(10)
```
